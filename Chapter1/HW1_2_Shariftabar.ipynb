**Problem 3:** Create a bar chart using Matplotlib to visualize the following data: Monthly sales for a store (January to May). Use appropriate labels and titles for the chart.

````python
import matplotlib.pyplot as plt
import numpy as np

labels = ["January", "February", "March", "April", "May"]
values = [100, 150, 182, 264, 310]
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color='magenta')
plt.xlabel('Month')
plt.ylabel('Sale')
plt.title('Bar Chart of Monthly Sales')

for i, v in enumerate(values):
  plt.text(i, v+1, str(v), ha='center', va='bottom')

plt.show()
````

**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
import pandas as pd

data = {
    "Name" : ["Maryam", "Nima", "Kian", "Sara", "Amin", "Pouya"],
    "Score" : [54, 73, 91, 86, 49, 80],
}

df = pd.DataFrame(data)
df['Grade'] = pd.cut(df['Score'], bins=[0, 70, 80, 90, float('inf')], labels=['D', 'C', 'B', 'A'])

print("Students lists with Grades = \n", df)
````

**Problem 11:** You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.


*   Dataset:

```
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
```
*  Tasks:
1. Calculate and display the total sales for each product over the two-year period.

2. Calculate and display the average monthly sales for each product.

3. Find the month with the highest sales for each product, and display the product and the sales value for that month.

4. Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

5. Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

6. Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.


````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

# Task-1
total_sales = sales_df [['Product A Sales', 'Product B Sales', 'Product C Sales']].sum()
print('Task1:\nTotal sales for each product =\n', total_sales, '\n')

# Task-2
average_sales = sales_df [['Product A Sales', 'Product B Sales', 'Product C Sales']].mean()
print('\nTask2:\nAverage monthly sales for each product =\n', average_sales, '\n')

# Task-3
max_sale_of_A = sales_df.loc[sales_df['Product A Sales'].idxmax()]['Month']
max_sale_value_of_A = sales_df['Product A Sales'].max()

max_sale_of_B = sales_df.loc[sales_df['Product B Sales'].idxmax()]['Month']
max_sale_value_of_B = sales_df['Product B Sales'].max()

max_sale_of_C = sales_df.loc[sales_df['Product C Sales'].idxmax()]['Month']
max_sale_value_of_C = sales_df['Product C Sales'].max()

print(f"\nTask3:\nHighest sales month of Product A is {max_sale_of_A.strftime('%B %Y')} with sales of {max_sale_value_of_A}")
print(f"Highest sales month of Product B is {max_sale_of_B.strftime('%B %Y')} with sales of {max_sale_value_of_B}")
print(f"Highest sales month of Product C is {max_sale_of_C.strftime('%B %Y')} with sales of {max_sale_value_of_C}\n")

# Task-4
jan_2022_sales = sales_df.loc[12, ['Product A Sales', 'Product B Sales', 'Product C Sales']]
dec_2022_sales = sales_df.loc[23, ['Product A Sales', 'Product B Sales', 'Product C Sales']]

percentage_change = ((dec_2022_sales - jan_2022_sales) * 100) / jan_2022_sales
print('\nTask4:\nPercentage Change Sales of January to December 2022 =\n ', percentage_change)

# Task-5
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.grid(True)
plt.plot(sales_df['Month'], sales_df['Product A Sales'], c='b', label='A')
plt.plot(sales_df['Month'], sales_df['Product B Sales'], c='r', label='B')
plt.plot(sales_df['Month'], sales_df['Product C Sales'], c='y', label='C')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for Each Product')
plt.legend(loc='upper left')
print('\nTask5:\n')
plt.show()

# Task-6
correlation_of_AB = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])
print(f"\nTask6:\nCorrelation between Product A and Product B sales = {correlation_of_AB}")
````
